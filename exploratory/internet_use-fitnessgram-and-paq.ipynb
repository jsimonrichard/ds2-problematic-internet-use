## Part 1: Data Visualization

__Author:__ Chase Ivancic

__Dataset:__ train.csv

__Columns:__ Internet Use, FitnessGram, Physical Activity Questionnaire (PAQ)

### __Internet Use:__ Number of hours of using computer/internet per day
- __PreInt_EduHx-Season:__ Season of participation (string)
> Spring, Summer, Fall, Winter
- __PreInt_EduHx-computerinternet_hoursday:__ Hours of using computer/internet (int)
> 0 (Less than 1h/day)  
> 1 (Around 1h/day)  
> 2 (Around 2hs/day)  
> 3 (More than 3hs/day)  

### FitnessGram Vitals & Treadmill: Measurement of cardiovascular fitness assessed using the NHANES treadmill protocol
- __Fitness_Endurance-Season:__ Season of participation (string)
> Spring, Summer, Fall, Winter
- __Fitness_Endurance-Max_Stage:__ Maximum Treadmill stage reached (int)
- __Fitness_Endurance-Time_Mins:__ Exact time completed: Minutes (int)
- __Fitness_Endurance-Time_Sec:__ Exact time completed: Seconds (int)

### FitnessGram Child: Health related physical fitness assessment measuring five differen parameters including aerobic capacity, muscular strength, muscular endurance, flexibility, and body composition
- __FGC-Season:__ Season of participation (string)
> Spring, Summer, Fall, Winter
- __FGC-FGC_CU:__ Curl up total (int)
- __FGC-FGC_CU_Zone:__ Curl up fitness zone (int)
> 0 - Needs improvement  
> 1 - Healthy fitness zone
- __FGC-FGC_GSND:__ Grip strength total (nondominant hand) (float)
- __FGC-FGC_GSND_Zone:__ Grip strength fitness zone (nondominant hand) (int)
> 1 - Weak  
> 2 - Normal  
> 3 - Strong
- __FGC-FGC_GSD:__ Grip strength total (dominant hand) (float)
- __FGC-FGC_GSD_Zone:__ Grip strength fitness zone (dominant hand) (int)
> 1 - Weak  
> 2 - Normal  
> 3 - Strong
- __FGC-FGC_PU:__ Push up total (int)
- __FGC-FGC_PU_Zone:__ Push up fitness zone (int)
> 1 - Weak  
> 2 - Normal  
> 3 - Strong
- __FGC-FGC_SRL:__ Sit & Reach total (left side) (float)
- __FGC-FGC_SRL_Zone:__ Sit & Reach fitness zone (left side) (int)
> 0 - Needs improvement  
> 1 - Healthy fitness zone
- __FGC-FGC_SRR:__ Sit & Reach total (right side) (float)
- __FGC-FGC_SRR_Zone:__ Sit & Reach fitness zone (right side) (int)
> 0 - Needs improvement  
> 1 - Healthy fitness zone
- __FGC-FGC_TL:__ Trunk lift total (int)
- __FGC-FGC_TL_Zone:__ Trunk lift fitness zone (int)
> 0 - Needs improvement  
> 1 - Healthy fitness zone
    
### Physical Activity Questionnaire: Information about childrens' participation in vigorous activities over the last 7 days
- __PAQ_A-Season:__ Season of participation (string)
> Spring, Summer, Fall, Winter
- __PAQ_A-PAQ_A_Total:__ Activity Summary Score (adolescents) (float)
- __PAQ_C-Season:__ Season of participation (string)
> Spring, Summer, Fall, Winter
- __PAQ_C-PAQ_C_Total:__ Activity Summary Score (Children) (float)
